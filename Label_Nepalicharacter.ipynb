```python
paths = r"D:\Nepali_character_Recognisation\Nepali_Character_dataset"
dir_list = os.listdir(paths)
# print("Files and directories in",paths,"':")
sorted_dir_list = sorted(dir_list)
total = 0
actual_path_list = []
for dir in sorted_dir_list:
    sub_path =os.path.join(paths,dir)
    # print(sub_path)
    # sub_dirlist = os.listdir(sub_path)
    # print(sub_dirlist)
    if os.path.isfile(sub_path):
        continue
    else:
        sub_dirlist = os.listdir(sub_path)
        sorted_sub_dirlist = sorted(sub_dirlist)
        for files in sorted_sub_dirlist:
            actual_path = os.path.join(sub_path,files)
            actual_path_list.append(actual_path)
            total +=1
print(f"Total Data: {total}")
actual_path_list.sort()
```
